The first step of the cleansing process is to identify the types of anomalies that reduce the data quality. Data cleaning is done by checking the data using some validation rules that are pre-specified, and then creating a report of the quality of the data and its problems. 
Data Anomalies can be any of the following three categories:

1. **Syntactic Anomalies**: 
describe characteristics concerning the format and values used for representation of the entities. Syntactic anomalies such as: lexical errors, domain format errors, syntactical error and irregularities.
2. **Semantic Anomalies**: 
hinder the data collection from being a comprehensive and non-redundant representation of the mini-world. These types of anomalies include: Integrity constraint violations, contradictions, duplicates and invalid tuples
3. **Coverage Anomalies**: 
decrease the amount of entities and entity properties from the mini-world that are represented in the data collection. Coverage anomalies are categorized as: missing values and missing tuples.

Lets fix null value of Origin regions first

```python
#from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
R = 6378.0

# This function will compare origin points with null origin points to find closest point and return correct region.

def region_na(originlat,originlong):  
    lat1 = radians(originlat)     #origin lattitude of null origin region
    lon1 = radians(originlong)    #origin lattitude of null origin region
    min_dist=[]
    for index, row in delivery_details.iterrows():
        if  str(row['Origin Region']) !='nan':  # for all non null origin region values
            destlong=row['Origin Longitude']  #origin longitude
            destlat=row['Origin Latitude']    #origin lattitude
            lat2 = radians(destlat)
            lon2 = radians(destlong)
            dlon = lon2 - lon1
            dlat = lat2 - lat1
            a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
            c = 2 * atan2(sqrt(a), sqrt(1 - a))
            distance = R * c   # calculate distance of null region origin point from all other region points 
            min_dist.append(distance) # append the distance in min_dist list
            if distance==min(min_dist):  # checking for minimum distance
                min_region = row['Origin Region']  #set min_region with region value of point least farther from null origin region point 
    return min_region

delivery_details.loc[delivery_details['Origin Region'].isnull(),'Origin Region'] = delivery_details[delivery_details['Origin Region'].isnull()].apply(lambda x: region_na(x['Origin Latitude'], x['Origin Longitude'] ), axis=1)
delivery_details
```

Checking for wrong origin region points and correcting them

According to kmean algorithm we calculate distances of all points from mean value point of each cluster so as to accurately place points in correct cluster region. 
Using this concept substitute correct region to wrong origin region values

```python
latitude = delivery_details.groupby('Origin Region', as_index=False)['Origin Latitude'].mean()  # calculating origin latitude mean for each region
longitude = delivery_details.groupby('Origin Region', as_index=False)['Origin Longitude'].mean() # calculating origin longitude mean for each region
Mean_lat_long = pd.merge(latitude,longitude, on="Origin Region") # creating a datframe Mean_lat_long with origin region, mean origin latitude, and mean origin longitude

#from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
R = 6378.0

# This function will compare destination point with mean point of each origin cluster region to find closest point to return correct region.

def region_fix(originlat,originlong):
    lat1 = radians(originlat)  # origin latitude
    lon1 = radians(originlong) #origin longitude
    min_dist=[]
    for index, row in Mean_lat_long.iterrows():
#         if  str(row['Origin Region']) !='nan':
            destlong=row['Origin Longitude']  # mean longitude 
            destlat=row['Origin Latitude']    #mean latitude
            lat2 = radians(destlat)
            lon2 = radians(destlong)
            dlon = lon2 - lon1
            dlat = lat2 - lat1
            a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
            c = 2 * atan2(sqrt(a), sqrt(1 - a))
            distance = R * c    # calculating distance of all origin points from mean origin region point of each cluster
            min_dist.append(distance) # append the distances in min_dist list
            if distance==min(min_dist): # checking for minimum distance
                min_regionval = row['Origin Region'] #set min_regionval with region value with least distant mean origin point from all other origin point 
    return min_regionval


delivery_details['Origin Region'] = delivery_details.apply(lambda x: region_fix(x['Origin Latitude'], x['Origin Longitude'] ), axis=1)
delivery_details
```

## checking for journey distance

```python
#from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
R = 6378.0

# function return distance between destination point and origin point

def distance(originlat,originlong,destlat,destlong): #ref1
    lat1 = radians(originlat)  # origin latitude
    lon1 = radians(originlong) # origin longitude
    lat2 = radians(destlat)  # destination latitude
    lon2 = radians(destlong) # destination longitude

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c  # distance between destination and origin points
    return round(distance,3) # rounding distance upto 3 decimal values


delivery_details['Journey Distance'] = delivery_details.apply(lambda x: distance(x['Origin Latitude'], x['Origin Longitude'],x['Destination Latitude'],x['Destination Longitude']), axis=1) #ref2
delivery_details
# delivery_details.to_csv('out.csv')

```

Above data shows that there are y-m-d format, y-d-m format, 30th feb date, 31st april and 31st june. Thus they are required to be converted into a single format and wrong dates need to be fix.

Converting all dates to proper Y-m-d format

```python
#import datetime

# function will convert departure date into y-m-d format

def date_extract(ddate):
    if str(pd.to_datetime(ddate, format='%Y-%m-%d', errors='coerce'))!='NaT':  # will check for Y-m-d format Ref3
        new_date = ddate
    elif str(pd.to_datetime(ddate, format='%Y-%d-%m', errors='coerce'))!='NaT': # will check for Y-d-m format
        year = ddate.split('-')[0].strip()  # year will have year value of date that is 0th index 
        month = ddate.split('-')[1].strip() # month will have date value of date that is 1st index
        date = ddate.split('-')[2].strip() # date will have month value of date that is 2nd index
        new_date = year + '-' + date + '-' + month  # we will interchange date with month value and store in new_date column
    elif ddate == '2018-02-30':
        new_date = '2018-02-03'  # will correct feb date
    elif ddate == '2018-04-31':
        new_date = '2018-05-01' # will correct april value
    elif ddate == '2018-06-31':
        new_date = '2018-07-01' # will correct june 
    return new_date.strip()  # to get the departure dates y-m-d format


delivery_details['Departure Date']=delivery_details['Departure Date'].apply(date_extract)
delivery_details
# delivery_details.to_csv('out.csv')
```

Calculating Speeds of drones to correct travel time 

```python
# calculating z values and beta values for each drone type.
z_drone = []
b_drone = []
for j in range(1,4):  # to iterate 3 drone types from 0 to 1
    z_mean = 0
    b_mean = 0
    z=[]
    d2_dt1 = delivery_details[delivery_details['Drone Type']==j] [['Journey Distance','Travel Time']].reset_index() #temp dataframe for each drone type reseting its index having journey distance and travel time as columns

    for i in range(0,len(d2_dt1)-1):    
        d2 = d2_dt1['Journey Distance'][i]   # taking n row Journey distance (n is 0 to length of frame-1)
        d1 = d2_dt1['Journey Distance'][i+1]  # taking n+1 row Journey distance
        t2 = d2_dt1['Travel Time'][i]       # taking n row Travel Time
        t1 = d2_dt1['Travel Time'][i+1]     # taking n+1 row Travel Time
        if t2 == t1 or d2== d1:
            temp =0
        else:
            temp=(d2-d1)/(t2-t1)      # calculating z with given formula
        if not np.isnan(temp):
            z.append(temp)           # appending all z values to z list

    z_mean=np.mean (z)              # calculating mean of z values for each drone  Ref-5
    
#    print(j,z_mean)

    b=[]
    for i in range(0,len(d2_dt1)):
        d1 = d2_dt1['Journey Distance'][i]  
        t1 = d2_dt1['Travel Time'][i]
        temp_b= d1-(z_mean*(t1-10))  # calculating beta for all rows
        if not np.isnan(temp_b):
            b.append(temp_b)   #appending beta for each row in b list

    b_mean=np.mean(b)   # Calculating mean beta for each drone type
   # print(j,b_mean)
    z_drone.append(z_mean)  # creating a list for 3 z values 
    b_drone.append(b_mean) # creating a list for 3 beta values
print(z_drone)
print(b_drone)
```

```python
def fix_null_drone_type(distance, time): # for fixing null drone types
    time_list=[]
    correct_time={}
    for i in range (3):
        correct_time[i+1] = 10 + ((distance-b_drone[i])/z_drone[i])  # create a dictionary with drone type as key and travel time as values
        time_diff = abs(time - correct_time[i+1])  # storing diff of all 3 drone type time in time_diff
        time_list.append(time_diff)  #appending these times to time list
        if time_diff == min(time_list):  # checking for the minimum time out of 3 drone type 
            dtype=i+1    # drone type will be minimum time associated drone type
    return dtype
    
delivery_details.loc[delivery_details['Drone Type'].isnull(),'Drone Type'] = delivery_details[delivery_details['Drone Type'].isnull()].apply(lambda x : fix_null_drone_type(x['Journey Distance'], x['Travel Time'] ), axis = 1)
delivery_details
```

```python
def fix_null_travel_time(distance, d_type):  # for fixing null travel time
    d_type=int(d_type)
    correct_time = 10 + ((distance-b_drone[d_type-1])/z_drone[d_type-1]) # calculating correct travel time as we know the drone types are correct
    return correct_time
    
delivery_details.loc[delivery_details['Travel Time'].isnull(),'Travel Time'] = delivery_details[delivery_details['Travel Time'].isnull()].apply(lambda x : fix_null_travel_time(x['Journey Distance'], x['Drone Type'] ), axis = 1)
delivery_details
```

```python
def fix_travel_time(distance, time, dronetype):
    time_list={}
    correct_time={}
    dronetype=int(dronetype)
    for i in range (3):
        correct_time[i+1] = 10 + ((distance-b_drone[i])/z_drone[i]) # create a dictionary with drone type as key and travel time as values
        time_diff = abs(time - correct_time[i+1])            # storing diff of all 3 drone type time in time_diff
        time_list[i+1]=time_diff        # storing all 3 differences in time_list
    min_time_dif=min(time_list.values())   # storing minimum difference in min_time_dif
    if min_time_dif< 0.1:  # if the diff is very less
        return time      # time will be original time 
    else:
        return correct_time[drontype]  # else time will be time associated with the drone type
        
    
delivery_details['Travel Time'] = delivery_details.apply(lambda x : fix_travel_time(x['Journey Distance'], x['Travel Time'],x['Drone Type'] ), axis = 1)
delivery_details
```

```python
def fix_drone_type(distance, time, dronetype):
    time_list={}
    correct_time={}
    for i in range (3):
        correct_time[i+1] = 10 + ((distance-b_drone[i])/z_drone[i]) # create a dictionary with drone type as key and travel time as values
        time_diff = abs(time - correct_time[i+1]) # storing diff of all 3 drone type time in time_diff
        time_list[i+1]=time_diff # storing all 3 differences in time_list
    min_time_dif=min(time_list.values())   # storing minimum difference in min_time_dif
    if min_time_dif<0.1:   # if the diff is very less
        for k,v in time_list.items():
            if v==min_time_dif:
                return k       # it will return drone type associated with minimum difference travel time 
            else :
                return dronetype  # else it will return original drone value
        
delivery_details['Drone Type']=delivery_details.apply(lambda x : fix_drone_type(x['Journey Distance'], x['Travel Time'],x['Drone Type'] ), axis = 1)
delivery_details
```

```python
# function predict delivery fare values depending on other variables
def delivery_fare(dronetype,posttype,jrnydist,traveltime,packagewt):
    y= m1*(dronetype)+m2*(posttype)+m3*(jrnydist)+m4*(traveltime)+m5*(packagewt)+c # predicted value
    return y

delivery_details['New Fare'] = delivery_details.apply(lambda x: delivery_fare(x['Drone Type'], x['Post Type'],x['Journey Distance'], x['Travel Time'],x['Package Weight']), axis = 1)
delivery_details
```

```python
def delivery_fare_diff(oldfare,newfare): # function to impute null value for delivery fare
    if oldfare == 0:  #if oldfare is 0
        return newfare   # fare will be predicted
    else:
        return oldfare  # else original fare

delivery_details['Delivery Fare'] = delivery_details.apply(lambda x: delivery_fare_diff(x['Delivery Fare'], x['New Fare']), axis = 1)
delivery_details
```
